# Task 01: Data Visualization – Distribution of Age in the Titanic Dataset

In this task, I performed data visualization using Python libraries such as Seaborn,Pandas and Matplotlib.

I used the Titanic dataset available in Seaborn and selected the **Age** column as a continuous variable. A histogram was created to visualize the distribution of passenger ages.

The visualization helped in understanding how passenger ages were distributed across different age groups. It showed that most passengers were young adults, while fewer passengers belonged to very young or older age groups.

Bar Chart (Categorical Variable)

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

sns.countplot(x='sex', data=titanic)
plt.title("Gender Distribution of Titanic Passengers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Histogram (Continuous Variable)

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

sns.histplot(titanic['age'], bins=20)
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
